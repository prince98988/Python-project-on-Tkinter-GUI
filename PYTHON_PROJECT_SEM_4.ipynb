```python
from tkinter import * 
import numpy as np
import math
import pandas as pd
import os
from functools import partial  
import random as rd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

#please don't use keyboard for calculator

root = Tk() 
root.title('21 CENTURY\'s CALCULATOR' ) 
expression=""
e1=""
l1=[];l=0;running=False; counter=-1;p=[];x=""

def input_number(number, equation):
    global expression
    global e1
    global l
    global l1
    global x
    
    if number=="HISTORY":
        if os.path.exists("HISTORY.txt"):
            os.remove("HISTORY.txt")
        return
    
    if number=="c" :
        if expression!="":
             
            if (e1[len(e1)-3:]=="sin" or e1[len(e1)-3:]=="tan" or e1[len(e1)-3:]=="cos" or e1[len(e1)-1:]=="√") and x!="":
                x=x[0:len(x)-1]
            else:
                e1=e1[0:len(e1)-l1[len(l1)-1]];l1.pop()
            expression=expression[0:len(expression)-l]
            equation.set(expression)
            return
        else:
            return
    try:
        if len(e1)>0:
            if e1[len(e1)-1:]=="√":
                
                if  str(number).isalnum():
                    x=x+str(number)
                    expression = expression + str(number)
                    equation.set(expression)
                    l=1;l1.append(1)
                    return
                else:
                    e1=e1[0:len(e1)-1]+str(math.sqrt(int(x)))
                    x=""

        if len(e1)>2:

            if e1[len(e1)-3:]=="sin":
                if (number=="-" and x=="") or str(number).isalnum():
                    x=x+str(number)
                    expression = expression + str(number)
                    equation.set(expression)
                    l=1;l1.append(1)
                    return
                else:
                    t=math.radians(int(x))
                    e1=e1[0:len(e1)-3]+str(np.sin(t))
                    x=""
            elif e1[len(e1)-3:]=="cos":
                if (number=="-" and x=="") or str(number).isalnum():
                    x=x+str(number)
                    expression = expression + str(number)
                    equation.set(expression)
                    l=1;l1.append(1)
                    return
                else:
                    t=math.radians(int(x))
                    e1=e1[0:len(e1)-3]+str(np.cos(t))
                    x=""
            elif e1[len(e1)-3:]=="tan":
                if (number=="-" and x=="") or str(number).isalnum():
                    x=x+str(number)
                    expression = expression + str(number)
                    equation.set(expression)
                    l=1;l1.append(1)
                    return
                else:
                    t=math.radians(int(x))
                    e1=e1[0:len(e1)-3]+str(np.tan(t))
                    x=""
                    
    except:
        pass
    
    if number=="^":
        e1=e1+"**"
        l=1;l1.append(2)
    elif number=="π":
        e1=e1+str(math.pi)
        l1.append(len(str(math.pi)))
    elif number=="e":
        e1=e1+"2.7182818284590" 
        l1.append(len("2.7182818284590"))
    else:
        e1=e1+str(number)
        l1.append(1)
    expression = expression + str(number)
    l=1
    equation.set(expression)
def evaluate(equation):
    global expression
    global e1,x
    try:
        if e1.find("√")!=-1:
            e1=e1[0:len(e1)-1]+str(math.sqrt(int(x)))
            x=""
        elif e1.find("sin")!=-1 :
            k=math.radians(int(x))
            e1=e1[0:len(e1)-3]+str(np.sin(k))
            x=""
        elif e1.find("cos")!=-1 :
            k=math.radians(int(x))
            e1=e1[0:len(e1)-3]+str(np.cos(k))
            x=""
        elif e1.find("tan")!=-1 :
            k=math.radians(int(x))
            e1=e1[0:len(e1)-3]+str(np.tan(k))
            x=""
            
        if e1!="":
            result = str(eval(e1))
        else:
            result=str(eval(equation.get()))
        equation.set(result)
        t=expression
        expression = result
        e1=result
        
        fo=open("HISTORY.txt","a+")
        fo.write(t+" = "+result+"\n")
        fo.close()
        
            
        
    except:
        pass

def clear_input_field(equation):
    global expression
    global e1
    expression = ""
    e1=""
    equation.set("0")

def calculation():
    equation = StringVar()
    e1 = Entry(root,bd =5,width=200,textvariable=equation)
    e1.grid(row=0, column=0,columnspan=4,padx=5,pady=10,ipady=3)
    equation.set("0")
    
    _1 = Button(root, text='1', fg='white', bg='#d1e0e0', bd=0,command=lambda: input_number(1, equation), height=2, width=7).grid(row=6, column=0)
    _2 = Button(root, text='2', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(2, equation), height=2, width=7).grid(row=6, column=1)
    _3 = Button(root, text='3', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(3, equation), height=2, width=7).grid(row=6, column=2)
    _4 = Button(root, text='4', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(4, equation), height=2, width=7).grid(row=5, column=0)
    _5 = Button(root, text='5', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(5, equation), height=2, width=7).grid(row=5, column=1)
    _6 = Button(root, text='6', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(6, equation), height=2, width=7).grid(row=5, column=2)
    _7 = Button(root, text='7', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(7, equation), height=2, width=7).grid(row=4, column=0)
    _8 = Button(root, text='8', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(8, equation), height=2, width=7).grid(row=4, column=1)
    _9 = Button(root, text='9', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(9, equation), height=2, width=7).grid(row=4, column=2)
    _0 = Button(root, text='0', fg='white', bg='#d1e0e0', bd=0, command=lambda: input_number(0, equation), height=2, width=7).grid(row=7, column=0)
    plus = Button(root, text='+', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('+', equation), height=2, width=7).grid(row=2, column=3)
    minus = Button(root, text='-', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('-', equation), height=2, width=7).grid(row=3, column=3)
    multiply = Button(root, text='*', fg='white', bg='#b3cccc', bd=0, command=lambda:  input_number('*', equation), height=2, width=7).grid(row=4, column=3)
    divide = Button(root, text='/', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('/', equation), height=2, width=7).grid(row=5, column=3)
    equal = Button(root, text='=', fg='white', bg='#b3cccc', bd=0, command=lambda: evaluate(equation), height=2, width=7).grid(row=6, column=3)
    clear = Button(root, text='CLEAR', fg='white', bg='#b3cccc', bd=0, command=lambda: clear_input_field(equation), height=2, width=7).grid(row=2, column=0)
    square=  Button(root, text='^', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('^',equation), height=2, width=7).grid(row=2, column=4)
    square=  Button(root, text='√', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('√',equation), height=2, width=7).grid(row=3, column=4)
    c1=   Button(root, text='(', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('(',equation), height=2, width=7).grid(row=2, column=1)
    c2=   Button(root, text=')', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number(')',equation), height=2, width=7).grid(row=3, column=1)
    pi= Button(root, text='π', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('π',equation), height=2, width=7).grid(row=2, column=2)
    e=  Button(root, text='e', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('e',equation), height=2, width=7).grid(row=3, column=2)
    sin=  Button(root, text='sin', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('sin',equation), height=2, width=7).grid(row=4, column=4)
    cos=  Button(root, text='cos', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('cos',equation), height=2, width=7).grid(row=5, column=4)
    tan=  Button(root, text='tan', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('tan',equation), height=2, width=7).grid(row=6, column=4)
    c=  Button(root, text='C', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('c',equation), height=2, width=7).grid(row=3, column=0)
    history=  Button(root, text='CLEAR HISTORY', fg='white', bg='#b3cccc', bd=0, command=lambda: input_number('HISTORY',equation), height=2, width=15).grid(row=7, column=4)
    
    
    
```

```python
def history():
    file = Toplevel(root)
    if os.path.exists("HISTORY.txt"):
        f=open("HISTORY.txt","r")
        k="".join(["->"+i for i in f])
        labelframe = LabelFrame(file, text="HISTORY")
        labelframe.pack(fill="both", expand="yes")
        left = Label(labelframe, text=k)
        left.pack()
    else:
        labelframe = LabelFrame(file, text="HISTORY")
        labelframe.pack(fill="both", expand="yes")
        left = Label(labelframe, text="EMPTY")
        left.pack()
    
    
def counter_label(label,end): 
    
    def count(): 
        global running,counter
       
        if running: 
            global counter 
            if counter==-1:             
                display="Starting..."
            else: 
                display=str(counter) 
  
            label['text']=display  
            label.after(1000, count)  
            counter += 1
            if counter==end:
                counter=-1
                running=False
                label['text']="UPGRADE YOUR SELF"
    count()  

    
def check(a,b,c,d,e,f,g,h,k,l,name,age,pa):
    check = Toplevel(root)
    global p,running
    running=False
    labelResult = Label(check,font=('calibre', 10, 'bold'))  
    labelResult.grid(row=0, padx=15, pady=10)  
    z=[a.get(),b.get(),c.get(),d.get(),e.get(),f.get(),g.get(),h.get(),k.get(),l.get()]
    z=np.array(z)
    p=np.array(p)
    y=list(abs(z-p)<0.15)
    t=y.count(True)
    z=[str(i) for i in p]
    labelResult.config(text="YOUR RESUT IS :"+str(t))  
    lab = Label(check)  
    lab.grid(row=2 ) 
    lab.config(text="CORRECT ANSWER  IS :"+"\n"+"\n".join(z)) 
    dict={"name":[name.get()],"Age":[age.get()],"Password":[pa.get()],"result":[t],"date":[str(datetime.today().strftime('%Y-%m-%d'))]}
    df=pd.DataFrame(dict)
    if os.path.exists("result.csv"):
        fo=open("result.csv","a")
        df.to_csv("result.csv", mode='a',header=False)
        fo.close()
    else:
        fo=open("result.csv","a")
        df.to_csv("result.csv", mode='a')
        fo.close()
        
 
def result():
    result = Toplevel(root)
    global show
    Name=StringVar() ;psw=StringVar();
    x=Label(result, text="RESULT",font=('calibre', 10, 'bold'), padx=30, pady=20,).grid(row=0)
    
    Label(result, text='userName').grid(row=2) 
    Label(result, text='password').grid(row=4)
    e1 = Entry(result,textvariable = Name)
    e2 = Entry(result,textvariable = psw,show="*")
    
    e1.grid(row=2, column=1) 
    e2.grid(row=4, column=1) 
    show = partial(show,Name,psw)
    R1 = Radiobutton(result, text="SEE YOUR INCREMENT", command=show).grid(row=13)
```

```python
def show(name,psw):
    show=Toplevel(root)
    Name=name.get()
    Pa=psw.get()
    df=pd.read_csv("result.csv")
    result=df[df['name']==Name]["result"]
    
    figure2 = plt.Figure(figsize=(5,4), dpi=100)
    ax2 = figure2.add_subplot(111)
    line2 = FigureCanvasTkAgg(figure2, show)
    line2.get_tk_widget().pack(side=LEFT, fill=BOTH)
    
    result.plot(kind='line', legend=True, ax=ax2, color='r',marker='o', fontsize=10)
    ax2.set_title('RESULT Vs. QUIZ')
def nquiz(name,age,pa):
    global running,check,p
    nquiz = Toplevel(root)
    
    labelResult = Label(nquiz,font=('calibre', 10, 'bold'))  
    labelResult.grid(row=0, padx=15, pady=10)  
    labelResult.config(text="WELCOME TO QUIZ "+name.get())  
    
    
    
    a3=IntVar();b3=IntVar();c3=IntVar();d3=IntVar();e33=IntVar();
    f3=IntVar();g3=DoubleVar();h3=DoubleVar();k3=IntVar();l3=IntVar()
    #-------------------------------
    a=rd.randint(1, 500)
    a11=rd.randint(1,500)
    a2='1 ADD :'+str(a)+"+"+str(a11)+" :"
    Label(nquiz, text=a2).grid(row=2) 
    #---------------------------------
    b=rd.randint(1, 500)
    b1=rd.randint(1,500)
    b2='2 ADD :'+str(b)+"+"+str(b1)+" :"
    Label(nquiz, text=b2).grid(row=3) 
    #---------------------------------
    c=rd.randint(1,500)
    c1=rd.randint(1,500)
    c2=a1='3 SUB :'+str(c)+"-"+str(c1)+" :"
    Label(nquiz, text=c2).grid(row=4)
    #---------------------------------
    d=rd.randint(1,500)
    d1=rd.randint(1,500)
    d2=a1='4 SUB :'+str(d)+"-"+str(d1)+" :"
    Label(nquiz, text=d2).grid(row=5)
    #---------------------------------
    e=rd.randint(1,500)
    e11=rd.randint(1,500)
    e2=a1='5 MUL :'+str(e)+"*"+str(e11)+" :"
    Label(nquiz, text=e2).grid(row=6)
    #---------------------------------
    f=rd.randint(1,500)
    f1=rd.randint(1,500)
    f2=a1='6 MUL :'+str(f)+"*"+str(f1)+" :"
    Label(nquiz, text=f2).grid(row=7)
    #---------------------------------
    g=rd.randint(1,500)
    g1=rd.randint(1,500)
    g2=a1='7 DIV :'+str(g)+"/"+str(g1)+" :"
    Label(nquiz, text=g2).grid(row=8)
    #---------------------------------
    h=rd.randint(1,500)
    h1=rd.randint(1,500)
    h2=a1='8 DIV :'+str(h)+"/"+str(h1)+" :"
    Label(nquiz, text=h2).grid(row=9)
    #---------------------------------
    k=rd.randint(1,500)
    k1=rd.randint(1,500)
    k2=a1='9 MOD :'+str(k)+"%"+str(k1)+" :"
    Label(nquiz, text=k2).grid(row=10)
    #---------------------------------
    l=rd.randint(1,500)
    l1=rd.randint(1,500)
    l2=a1='10 MOD :'+str(l)+"%"+str(l1)+" :"
    Label(nquiz, text=l2).grid(row=11)
    #---------------------------------
    
    e1 = Entry(nquiz,textvariable =a3)
    e2 = Entry(nquiz,textvariable = b3)
    e3 = Entry(nquiz,textvariable = c3)
    e4 = Entry(nquiz,textvariable = d3)
    e5 = Entry(nquiz,textvariable = e33)
    e6 = Entry(nquiz,textvariable = f3)
    e7 = Entry(nquiz,textvariable = g3)
    e8 = Entry(nquiz,textvariable = h3)
    e9 = Entry(nquiz,textvariable = k3)
    e10 = Entry(nquiz,textvariable = l3)
    e1.grid(row=2, column=1) 
    e2.grid(row=3, column=1) 
    e3.grid(row=4, column=1) 
    e4.grid(row=5, column=1) 
    e5.grid(row=6, column=1) 
    e6.grid(row=7, column=1) 
    e7.grid(row=8, column=1) 
    e8.grid(row=9, column=1) 
    e9.grid(row=10, column=1) 
    e10.grid(row=11, column=1) 
   
    
    label = Label(nquiz,font=('calibre', 10, 'bold'))  
    label.grid(row=1, column=2)  
    running=True;
    counter_label(label,200) 
    p=[a+a11,b+b1,c-c1,d-d1,e*e11,f*f1,round(g/g1,2),round(h/h1,2),k%k1,l%l1]
    check = partial(check,a3,b3,c3,d3,e33,f3,g3,h3,k3,l3,name,age,pa)
    R1 = Radiobutton(nquiz, text="SEE RESULT", command=check).grid(row=13)


def quiz():
    global nquiz
    quiz = Toplevel(root)
    name=StringVar() ;age=IntVar();pa=StringVar();var=IntVar()
    x=Label(quiz, text='ENTER YOUR DATA',font=('calibre', 10, 'bold'), padx=20, pady=20,).grid(row=0)
    
    Label(quiz, text='userName').grid(row=2) 
    Label(quiz, text='Age').grid(row=3)
    Label(quiz, text='password').grid(row=4)
    Label(quiz, text='ENTER CHOICE',font=('calibre',10,'normal')).grid(row=5)
    e1 = Entry(quiz,textvariable = name)
    e2 = Entry(quiz,textvariable = age)
    e3 = Entry(quiz,textvariable = pa,show="*")
    
    e1.grid(row=2, column=1) 
    e2.grid(row=3, column=1) 
    e3.grid(row=4, column=1) 

    nquiz = partial(nquiz,name,age,pa )
    R1 = Radiobutton(quiz, text="Start Quiz", variable=var, value=1,command=nquiz).grid(row=6)
   
    
```
